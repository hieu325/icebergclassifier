# Simple (Shallow) Neural Network base on max-median of image (LogLoss~0.5)

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_json('./data/processed/train.json')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn.model_selection import train_test_split
from keras.optimizers import Adam
from keras.regularizers import l1
```

```python
b1max=[np.max(np.asarray(b)) for b in train['band_1'].values]
b1med=[np.median(np.asarray(b)) for b in train['band_1'].values]
b2max=[np.max(np.asarray(b)) for b in train['band_2'].values]
b2med=[np.median(np.asarray(b)) for b in train['band_2'].values]
X_data = np.column_stack((b1max,b1med,b2max,b2med))
y_data = train['is_iceberg'].values
X_train, X_dev, y_train, y_dev = train_test_split(X_data, y_data, random_state=16, test_size=0.1)
```

```python
model=Sequential()

# adding neural network layers
model.add(Dense(units=20,input_dim=X_train.shape[1]))
model.add(Activation('relu'))
model.add(Dense(units=20))
model.add(Activation('relu'))
model.add(Dense(units=20))
model.add(Activation('relu'))
model.add(Dense(units=1,activation='sigmoid'))

model.compile(optimizer=Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0),loss='binary_crossentropy')
model.fit(X_train,y_train,epochs=1000,verbose=2, validation_data=(X_dev,y_dev));
```
